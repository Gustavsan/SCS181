```python
def Gproduct(a,b,N,B,k):
    if a == float("inf"):
        return b
    elif b == float("inf"):
        return a
    elif pow(a,2,N) == B and pow(b,2,N) == B:
        return False
    elif k == 0:
        return 1
    elif k == 1 and a == b:
        return a
    elif k > 1 and a == b:
        k -= 1
        if pow(a + b,1,N) == 0:
            return float("inf")
        elif math.gcd(a + b,N) != 1:
            return "N - composite GProduct"
        elif math.gcd(a + b ,N) == 1:
            for i in range (0,1001):
                ab = (pow((a*b + B)*(pow(a+b,-1,N)),1,N) + N*i)
                if pow(ab,2,N) == B:
                      return "N - composite GProduct"
                elif k == 1:
                    return ab
                else:
                    return Gproduct(ab,b,N,B,k-1)
    else:
        if pow(a + b,1,N) == 0:
            return float("inf")
        elif math.gcd(a + b,N) != 1:
            return "N - composite GProduct"
        elif math.gcd(a + b ,N) == 1:
            for i in range (0,1001):
                ab = (pow((a*b + B)*(pow(a+b,-1,N)),1,N) + N*i)
                if pow(ab,2,N) == B:
                      return "N - composite GProduct"
                elif k == 1:
                    return ab
                else:
                    return Gproduct(ab,b,N,B,k-1)
```
